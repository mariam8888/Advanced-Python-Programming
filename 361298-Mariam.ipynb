#**Load Image and Display**

```python
# Import the necessary libraries
from google.colab.patches import cv2_imshow
import cv2
# load the image and display
img = cv2.imread('cat.png', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

#**Convert Image into Numpy Array**

```python
import numpy
np_img = numpy.array(img)
print("Size of Numpy Array: ", np_img.shape)
print("Converted Numpy Array is as follows: \n", np_img)
```

#**Apply Reshaping**

```python
conv_img = cv2.resize(img, dsize=(28, 28))
conv_img = conv_img.reshape(28,28,3)
print("Size of Numpy Array: ", conv_img.shape)
print("Converted Numpy Array is as follows: \n", conv_img)
```

```python
from PIL import Image as im
# creating image object of
# above array
data = im.fromarray(np_img)
      
# saving the final output 
# as a PNG file
data.save('gfg_dummy_pic.png')
```
